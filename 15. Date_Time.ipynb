## Working with Date

```python
# Converting to datetime datatype

date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1. Extract Year

```python
date['date_year'] = date['date'].dt.year

date.head()
```

#### 2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
# For Extracting month name

date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

#### 3. Extracting Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# Extracting  day of week

date['date_DOW'] = date['date'].dt.day_of_week

date.head()
```

```python
# Extracting Day of Week - Name

date['date_DOW_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Is Weekend ?

date['date_is_weekend'] = np.where(date['date_DOW_name'].isin(['Sunday','Saturday']),1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 4. Extracting Week of the Year

```python
# date['date_week'] = date['date'].dt.weekday # This works in earlier pandas version

date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 5. Extracting Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 6. Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 7. Extracting Time Elapsed Between Date (Gap(Difference) between Two Dates)

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
# Extracting Elapsed Days

(today - date['date']).dt.days
```

```python
# Month Passed

np.round((today - date['date'])/np.timedelta64(1,'M'),0)
```

## Working With Time

```python
time
```

```python
time.info()
```

```python
# Converting to datetime datatype

time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

#### 1. Extracting Time Part

```python
time['time'] = time['date'].dt.time

time.head()
```

#### 2. Extracting Time Difference

```python
today - time['date']
```

```python
# Time Difference in Seconds

(today - time['date'])/np.timedelta64(1, 's')
```

```python
# Time Difference in Minutes

(today - time['date'])/np.timedelta64(1, 'm')
```

```python
# Time Difference in Hours

(today - time['date'])/np.timedelta64(1, 'h')
```
